```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 15, 7]

plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

# 2.Bar chart

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 5]

plt.bar(categories, values, color='skyblue')
plt.title("Bar Chart")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("titanic-data.csv")

# Scatter plot Age vs Fare
plt.figure(figsize=(8,6))
plt.scatter(df['Age'], df['Fare'], alpha=0.5)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs Fare')
plt.show()
```

## 4.Histogram

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("titanic-data.csv")

# Drop missing Age values to avoid errors
ages = df['Age'].dropna()

# Plot histogram
plt.figure(figsize=(8,6))
plt.hist(ages, bins=30, edgecolor='black', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.title('Age Distribution of Titanic Passengers')
plt.show()
```

# 7.Heatmap

Heatmaps plot numerical data and provide visualizations that reveal patterns, trends, and anomalies within datasets. They use color-coded grids to represent data values, making it easy to identify areas of high or low magnitude. This allows for quick understanding of relationships between variables and can be particularly useful for exploring large datasets.
