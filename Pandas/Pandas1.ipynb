```python
import pandas as pd

# Sample DataFrame with custom index
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [24, 27, 22]
}, index=['a', 'b', 'c'])

# Filter row with index label 'b'
print(df.loc['b'])
```
