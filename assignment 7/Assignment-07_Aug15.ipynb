### **Problem 3: Handling Missing Data and Imputation**  
1. Identify columns with **missing values**.
2. Impute the missing values:
   - For **numerical columns**, replace missing values with the **median**.
   - For **categorical columns**, replace missing values with the **most frequent** value.
3. Justify your choice of imputation methods and check if any missing data remains.

```python
import pandas as pd

df = pd.read_csv("train.csv")

missing_values = df.isnull().sum()
missing_df = missing_values[missing_values > 0].sort_values(ascending=False)

print("Columns with missing values:")
print(missing_df)
```

```python
import pandas as pd

df = pd.read_csv("train.csv")

num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include='object').columns

df[num_cols] = df[num_cols].fillna(df[num_cols].median())

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

missing_values_after = df.isnull().sum()
print("Missing values after imputation:")
print(missing_values_after[missing_values_after > 0])
```

```python
import pandas as pd

df = pd.read_csv("train.csv")

num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include='object').columns

df[num_cols] = df[num_cols].fillna(df[num_cols].median())

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

missing_values_after = df.isnull().sum()
print("Missing values after imputation:")
print(missing_values_after[missing_values_after > 0])

```

### **Problem 4: Numpy Array Operations**  
1. Convert the numerical columns (such as `LotArea`, `YearBuilt`, `GrLivArea`, `SalePrice`, etc.) into a **NumPy array**.
2. Perform the following NumPy operations on the array:
   - Compute the **mean** and **standard deviation** of each column.
   - Normalize the data using **min-max scaling** and **z-score normalization**.
   - Compute the **dot product** between two numerical columns of your choice.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("train.csv")

num_cols = ['LotArea', 'YearBuilt', 'GrLivArea', 'SalePrice']
num_data = df[num_cols].values

mean_vals = np.mean(num_data, axis=0)
std_vals = np.std(num_data, axis=0)

print("Mean of each column:")
for col, mean in zip(num_cols, mean_vals):
    print(f"{col}: {mean:.2f}")

print("\nStandard Deviation of each column:")
for col, std in zip(num_cols, std_vals):
    print(f"{col}: {std:.2f}")

min_vals = np.min(num_data, axis=0)
max_vals = np.max(num_data, axis=0)
min_max_scaled = (num_data - min_vals) / (max_vals - min_vals)
print("\nMin-Max Scaled Data (first 5 rows):")
print(min_max_scaled[:5])

z_score_normalized = (num_data - mean_vals) / std_vals
print("\nZ-score Normalized Data (first 5 rows):")
print(z_score_normalized[:5])

dot_product = np.dot(df['GrLivArea'], df['SalePrice'])
print(f"\nDot product of GrLivArea and SalePrice: {dot_product:.2f}")
```

### **Problem 6: Aggregation and Transformation**  
1. Perform a **groupby** operation on `Neighborhood` and calculate the **mean** and **median** for `SalePrice`.
2. Use the `apply()` function to create a custom transformation for `SalePrice` (e.g., scaling or transformation).
3. Aggregate multiple statistical values (e.g., **mean**, **median**, and **count**) for `SalePrice` grouped by `OverallQual`.

```python
import pandas as pd

df = pd.read_csv("train.csv")

neighborhood_stats = df.groupby('Neighborhood')['SalePrice'].agg(['mean','median']).reset_index()
print("Mean and Median SalePrice by Neighborhood:")
print(neighborhood_stats)

df['SalePrice_Log'] = df['SalePrice'].apply(lambda x: np.log1p(x))
print("\nFirst 5 rows with log-transformed SalePrice:")
print(df[['SalePrice','SalePrice_Log']].head())

overallqual_stats = df.groupby('OverallQual')['SalePrice'].agg(['mean','median','count']).reset_index()
print("\nMean, Median, and Count of SalePrice by OverallQual:")
print(overallqual_stats)
```
